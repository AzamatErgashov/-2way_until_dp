```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

# data

noise real dataga uxshashi uchun malumotga shovqin qushadi

```python
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0],data[1]
plt.scatter(x_data,y_data)
plt.show()
```

# demak malumotimiz usib boryapti biz bu malumotga klassik dasturchi sifatida yondashganimizda 2.0 teng 140lar bulyapti yani 70 barobarga kopayadi deb funksiya yozardim

```python
def classic_program(x): return x * 70

y_pred = [classic_program(x) for x in x_data]
```

```python
plt.scatter(x_data,y_data)
plt.plot(x_data,y_pred, 'r')
plt.show()
```

# manabu bizning classic dasturlashdagi qilgan funksiyamizning ishi

```python
def model_program(x, w): return w * x * 70
w = 1.3
y_pred = [model_program(x, w) for x in x_data]

plt.scatter(x_data,y_data)
plt.plot(x_data,y_pred, 'r')
plt.show()
```
